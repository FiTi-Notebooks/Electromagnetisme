```python
#####################################################################
#                                                                   #
#             EXECUTER CETTE CELLULE SANS LA MODIFIER               #
#                                                                   #
#####################################################################

# The FiTi Project : Fiches interactives de Travail individuel 
# https://github.com/FiTi-Notebooks
# Apache License 2.0

# Sorbonne Université
# TP L2 Ondes et electromagnetisme
# TP3 électrocinétique et filtrage
# Auteur : Arnaud Cassan

#####################################################################

import sys
import numpy as np
import matplotlib.pyplot as plt

# stop execution
def stop():
    """Arreter execution apres cette ligne.
    """
    sys.exit(0)
   
# common plot parameters
parameters = {'axes.labelsize': 15, 'axes.titlesize': 15, 
              'xtick.labelsize': 15, 'ytick.labelsize': 15,
              'figure.figsize': (12, 5)}
plt.rcParams.update(parameters)

# complex number i
global j
j = complex(0., 1.)

# argument of z
def mod(z):
    """Module du nombre complexe z.
    """
    return np.abs(z)

# module of z
def arg(z):
    """Argument du nombre complexe z.
    """
    return np.angle(z)

# create x values
def xvals(xmin, xmax, n):
    """Creation d'un tableau de valeurs pour x,
    uniformement reparties en log entre xmin et xmax.
    """
    return np.logspace(np.log10(xmin), np.log10(xmax), n, endpoint=True)

# plot modulus of H
def plot_mod(x, H, xrange=None, yrange=None):
    """Trace le module de H(w/wc).
    """
    plt.subplots()
    ax = plt.subplot()
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$\omega/\omega_c$')
    ax.set_ylabel(r'$G=\|H\|$')
    ax.plot(x, np.abs(H))
    if xrange: ax.set_xlim(*xrange)
    if yrange: ax.set_ylim(*yrange)
    plt.grid()
    plt.show()

# plot argument of H
def plot_arg(x, H, xrange=None, yrange=None):
    """Trace l'argument de H(w/wc).
    """
    plt.subplots()
    ax = plt.subplot(111)
    ax.set_xscale('log')
    ax.set_xlabel(r'$\omega/\omega_c$')
    ax.set_ylabel(r'$\phi=\arg(H)$')
    ax.plot(x, np.angle(H))
    if xrange: ax.set_xlim(*xrange)
    if yrange: ax.set_ylim(*yrange)
    plt.grid()
    plt.show()
    
# plot module and argument of H
def plot_modarg(x, H):
    """Trace le module et l'argument de H(w/wc).
    """
    plt.subplots(figsize=(12, 10))
    ax = plt.subplot(211)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$\omega/\omega_c$')
    ax.set_ylabel(r'$G=\|H\|$')
    ax.plot(x, np.abs(H))
    plt.grid()
    ay = plt.subplot(212, sharex=ax)
    ay.set_xscale('log')
    ay.set_xlabel(r'$\omega/\omega_c$')
    ay.set_ylabel(r'$\phi=\arg(H)$')
    ay.plot(x, np.angle(H))
    plt.grid()
    plt.show()
    
# plot function
def plot(x, F, xscale='linear', yscale='linear', title='Value'):
    """Trace la fonction F(w/wc).
    """
    plt.subplots()
    ax = plt.subplot(111)
    ax.set_xscale(xscale)
    ax.set_yscale(yscale)
    ax.set_xlabel(r'$\omega/\omega_c$')
    ax.set_ylabel(title)
    ax.plot(x, F)
    plt.grid()
    plt.show()


# plot function
def plot_exp(x, F, xscale='linear', yscale='linear', xrange=None, yrange=None, title='Value'):
    """Trace les donnees.
    """
    plt.subplots()
    ax = plt.subplot(111)
    ax.set_xscale(xscale)
    ax.set_yscale(yscale)
    ax.set_xlabel(r'Frequence (kHz)')
    ax.set_ylabel(title)
    ax.scatter(np.array(x), np.array(F))
    if xrange: ax.set_xlim(*xrange)
    if yrange: ax.set_ylim(*yrange)
    plt.grid()
    plt.show()
      
# built signal square signal
def carre(n):
    t = np.linspace(0., 2., 1000)
    f = 1.
    s = 0.
    for k in range(n):
        signal = np.sin(2. * (2. * k + 1.) * np.pi * f * t) / (2. * k + 1.)
        s += signal
        plt.plot(t, signal)
    plt.plot(t, s, c='k', lw=3)
    plt.show()
```

```python
# QUESTIONS 4.1.1.c) et d)
print("NIVEAU : FACILE")

# x = omega/omega_c : 
x = xvals(0.01, 100., 40) # (min, max, nombre de valeurs)


### EXEMPLE DE SCRIPT :

# definition de H
H = j * x / (1. + j * x)

# tracé du module et de l'argument de H
plot_modarg(x, H) 
```

```python
# QUESTIONS 4.1.1.c) et d)
print("NIVEAU : MOYEN")

# x = omega/omega_c : 
x = xvals(0.01, 100., 40) # (min, max, nombre de valeurs)


### EXEMPLE DE SCRIPT :

# définition de H
H = j * x / (1. + j * x)

# tracé du module de H
plot_mod(x, H) 

# tracé de l'argument de H
plot_arg(x, H) 
```

```python
# QUESTIONS 4.1.1.c) et d)
print("NIVEAU : DIFFICILE")

# x = omega/omega_c : 
x = xvals(0.01, 100., 40) # (min, max, nombre de valeurs)


### EXEMPLE DE SCRIPT :

# définition de H
H = j * x / (1. + j * x)

# calcul du module de H
G = mod(H)

# tracé de G
plot(x, G, xscale='log', yscale='log', title='G')

# calcul de l'argument de H
phi = arg(H)

# tracé de phi
plot(x, phi, xscale='log', title='phi')
```

```python
# QUESTION 4.1.2.h)


### EXEMPLE DE SCRIPT :

# données expérimentales
f = [0.1, 0.2, 0.5, 1., 2., 5., 10., 20., 50., 100.]
G = [0.05, 0.10, 0.24, 0.45, 0.71, 0.93, 0.98, 0.99, 1., 1.]
phi = [1.52, 1.47, 1.33, 1.11, 0.79, 0.38, 0.20, 0.10, 0.04, 0.02]

# tracé de G
plot_exp(f, G, xscale='log', yscale='log', xrange=[0.02, 200.], yrange=[0.01, 2.], title='G')

# tracé de phi
plot_exp(f, phi, xscale='log', xrange=[0.02, 200.], yrange=[0., 1.6], title='phi')
```

```python
# QUESTIONS 4.2.1.e)

# x = omega/omega_c : 
x = xvals(0.1, 10., 200) # (min, max, nombre de valeurs)


### EXEMPLE DE SCRIPT :

Q1 = 1.
Q2 = 10.
Q3 = 50.

H1 = 1. / (1. + j * Q1 * (x - 1. / x))
H2 = 1. / (1. + j * Q2 * (x - 1. / x))
H3 = 1. / (1. + j * Q3 * (x - 1. / x))


# tracé du module et de l'argument de H
plot_mod(x, H1, yrange=[0.01, 2.])
plot_mod(x, H2, yrange=[0.01, 2.])
plot_mod(x, H3, yrange=[0.01, 2.])
```
